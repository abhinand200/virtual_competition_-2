Filling null value

```python
plt.hist(df['Credit_History'])
plt.show()
```

```python
plt.hist(df['Loan_Amount_Term'])
plt.show()
```

```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
df['Married'].fillna(df['Married'].mode()[0],inplace=True)
df['Dependents'].fillna(df['Dependents'].mode()[0],inplace=True)
df['Self_Employed'].fillna(df['Self_Employed'].mode()[0],inplace=True)

df['Credit_History'].fillna(df['Credit_History'].median(),inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median(),inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].median(),inplace=True)
```

```python
df.isnull().sum()
```

Filling outier

```python
df.boxplot()
plt.xticks(rotation='vertical')
plt.show()
```

```python
for col in df.columns:
  if df[col].dtype != 'object':
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    upper=q3+1.5*iqr
    lower=q1-1.5*iqr
    df[col]=np.where(df[col]>upper,upper,np.where(df[col]<lower,lower,df[col]))
```

**Encordingt**

This columns are object type it an be convert to int(one hot encoding)

```python
df.dtypes
```

```python
df=pd.get_dummies(df,columns=['Gender','Married','Dependents','Education','Self_Employed','Property_Area'],dtype=int)
```

```python
df.head()
```

Label Encording

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Loan_Status']=le.fit_transform(df['Loan_Status'])
```

spliting data for train and test

```python
x = df.drop(columns=['Loan_Status', 'Loan_ID'], axis=1)
y = df['Loan_Status']
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

SVM

```python
from sklearn.svm import SVC
sc_clf=SVC(kernel='linear',random_state=42)
sc_clf.fit(x_train,y_train)
y_pred_svc=sc_clf.predict(x_test)
```

```python
y_pred_svc.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print('svc_accuracy is',accuracy_score(y_test,y_pred_svc))
print('svc_precision is',precision_score(y_test,y_pred_svc))
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn_clf=KNeighborsClassifier(n_neighbors=5)
knn_clf.fit(x_train,y_train)
y_pred_knn=knn_clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print('knn accuracy is',accuracy_score(y_test,y_pred_knn))
print('svc_precision is',precision_score(y_test,y_pred_knn))
```

DT

```python
from sklearn.tree import DecisionTreeClassifier
dt_clf=DecisionTreeClassifier()
dt_clf.fit(x_train,y_train)
y_pred_dt=dt_clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print('dt accuracy is',accuracy_score(y_test,y_pred_dt))
print('dt_precision is',precision_score(y_test,y_pred_dt))
```

RF

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf=RandomForestClassifier(n_estimators=100)
rf_clf.fit(x_train,y_train)
y_pred_rf=rf_clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print('rf accuracy is',accuracy_score(y_test,y_pred_rf))
print('rf_precision is',precision_score(y_test,y_pred_rf))
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_clf=LogisticRegression(random_state=42)
lr_clf.fit(x_train,y_train)
y_pred_lr=lr_clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print('lr accuracy is',accuracy_score(y_test,y_pred_lr))
print('lr_precision is',precision_score(y_test,y_pred_lr))
```

```python
print('lr accuracy is',accuracy_score(y_test,y_pred_lr))
print('rf accuracy is',accuracy_score(y_test,y_pred_rf))
print('dt accuracy is',accuracy_score(y_test,y_pred_dt))
print('knn accuracy is',accuracy_score(y_test,y_pred_knn))
print('svm accuracy is',accuracy_score(y_test,y_pred_svc))
```

```python
print('svm_precision is',precision_score(y_test,y_pred_svc))
print('lr_precision is',precision_score(y_test,y_pred_lr))
print('dt_precision is',precision_score(y_test,y_pred_dt))
print('knn_precision is',precision_score(y_test,y_pred_knn))
print('rf_precision is',precision_score(y_test,y_pred_rf))
```
